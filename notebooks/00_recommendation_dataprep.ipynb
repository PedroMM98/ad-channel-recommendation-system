```python
import pandas as pd
import numpy as np

```

```python
# Cargar el dataset de publicidad en redes sociales
# https://www.kaggle.com/datasets/jsonk11/social-media-advertising-dataset?resource=download 
df = pd.read_csv("Social_Media_Advertising.csv")
```

```python
df.head()
```

```python
# Para este ejemplo nos concentraremos en "Los Angeles" 
# y en el sector "Technology" para publicidad en Inglés
df = df[
    (df["Location"] == "Los Angeles")
     & (df["Customer_Segment"] == "Technology")
     & (df["Language"] == "English")
].reset_index(drop=True)
```

```python
df.to_csv("social_media_ads_filtered.csv", index=False)
```
